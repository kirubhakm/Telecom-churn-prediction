```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv('/content/customer_churn1 python.csv')
```

```python
X = data[['Monthly Charge', 'Tenure in Months', 'Total Charges', 'Contract']]
y = data['CustomerStatus']

```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
X['Contract'] = label_encoder.fit_transform(X['Contract'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier()
```

```python
clf = RandomForestClassifier()
```

```python
# Call the fit method with the training data
clf.fit(X_train, y_train) # Fit the model with the training data
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
data['PredictedChurn'] = clf.predict(X) # Use clf instead of model
```

```python
# Save predictions to CSV for Power BI
data[['Customer ID', 'PredictedChurn']].to_csv('churn_predictions.csv', index=False)
```
